A **virtual environment** in Python is a self-contained directory that holds a specific version of Python and all the packages you install for a particular project. It's like having a separate, isolated workspace for each of your Python projects.

**Why use virtual environments?**

*   **Dependency Management:** Different projects might require different versions of the same library. Without virtual environments, you could run into conflicts. Virtual environments ensure each project's dependencies are isolated.
*   **Cleanliness:** Keeps your global Python installation clean and free from project-specific packages.
*   **Reproducibility:** Makes it easy to share your project with others, as they can set up an identical environment by installing the listed dependencies.

**Commonly used tools:**

*   `venv` (built-in Python module)
*   `conda` (for Anaconda/Miniconda distributions)
*   `pipenv`
*   `poetry`

### `df.loc()`: Accessing data by label

`df.loc[]` is a powerful, label-based indexer used for selecting data from a Pandas DataFrame. It allows you to select rows and columns by their labels (names).

**Syntax:**

```python
df.loc[row_label_indexer, column_label_indexer]
```

**Key features:**

*   **Label-based:** You use the actual row and column names (labels) to make selections, not integer positions.
*   **Inclusive:** When slicing with labels (e.g., `start_label:end_label`), both the `start_label` and `end_label` are included in the selection.
*   **Boolean indexing:** Can be used with boolean arrays for conditional selection of rows.

**Examples:**

1.  **Select a single row by label:**
    ```python
    df.loc['row_name']
    ```

2.  **Select multiple rows by labels:**
    ```python
    df.loc[['row_name1', 'row_name2']]
    ```

3.  **Select a range of rows by labels (inclusive):**
    ```python
    df.loc['start_row_name':'end_row_name']
    ```

4.  **Select rows and specific columns by labels:**
    ```python
    df.loc[['row_name1', 'row_name2'], ['column_name1', 'column_name2']]
    ```

5.  **Select all rows and specific columns:**
    ```python
    df.loc[:, ['column_name1', 'column_name2']]
    ```

6.  **Conditional selection (boolean indexing):**
    ```python
    df.loc[df['column_name'] > 10]
    ```

7.  **Select specific rows conditionally and specific columns:**
    ```python
    df.loc[df['column_name'] == 'value', 'another_column']
    ```

**Syntax:**

### `pandas.DataFrame.pivot_table()`: `index`, `columns`, `values` Explained (Short)

These three parameters define how your data is grouped and summarized in a pivot table:

*   **`index`**: Specifies the column(s) whose unique values will form the **rows** of your pivot table. Think of these as your primary grouping categories that go down the side.

*   **`columns`**: Specifies the column(s) whose unique values will form the **columns** of your pivot table. These are your secondary grouping categories that go across the top.

*   **`values`**: Specifies the numerical column(s) that you want to **aggregate** (e.g., sum, average, count) within the cells of your pivot table, based on the `index` and `columns` groupings.

**In essence:** `index` defines the rows, `columns` defines the new columns, and `values` is the data being summarized at their intersections.

````python
# Top 6 countries
top_countries = df['job_country'].value_counts().head(6).index

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

# Empty dataframe to store results
df_job_and_country_salary = pd.DataFrame(index=top_countries, columns=job_titles)

# Fill values using groupby
for country in top_countries:
    for job in job_titles:
        median_salary = (
            df[(df['job_country'] == country) &
               (df['job_title_short'] == job)]
            ['salary_year_avg']
            .median()
        )
        df_job_and_country_salary.loc[country, job] = median_salary

# Plot
df_job_and_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
````

but we can use unstack 

What unstack() is

unstack() is a reshaping operation in pandas.
It moves one level of a row index into columns.

Conceptually:

row index level → column labels

When you run:

df.groupby(['job_country','job_title_short'])['salary_year_avg'].median()

the result is a Series with a MultiIndex:

job_country   job_title_short
USA           Data Analyst        85000
USA           Data Engineer      110000
USA           Data Scientist     120000
UK            Data Analyst        65000


What unstack() does
.unstack()

moves the second index level (job_title_short) into columns.

Result:

job_country   Data Analyst   Data Engineer   Data Scientist
USA              85000          110000           120000
UK               65000           90000            95000

````python
# Top 6 countries by job postings
top_countries = df['job_country'].value_counts().head(6).index

# Groupby equivalent of pivot_table
df_job_and_country_salary = (
    df.groupby(['job_country', 'job_title_short'])['salary_year_avg']
    .median()
    .unstack()
)

# Keep only top countries
df_job_and_country_salary = df_job_and_country_salary.loc[top_countries]

# Select specific job titles
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_job_and_country_salary = df_job_and_country_salary[job_titles]

# Plot
df_job_and_country_salary.plot(kind='bar')
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
````

### `pandas.DataFrame.reset_index()`: Reset the DataFrame Index

`DataFrame.reset_index()` is used to reset the index of a DataFrame back to the default integer index (0, 1, 2, ...). It converts the current index into one or more regular columns in the DataFrame.

**Full Syntax:**

```python
df.reset_index(level=None, *, drop=False, inplace=False, col_level=0, col_fill='', allow_duplicates=False, names=None)
```

**Key Parameters Explained:**

*   `level`: (Optional) Specifies which level(s) of a MultiIndex to reset. By default, it resets all levels.
*   `drop`: If `True`, the old index will *not* be added as a column to the DataFrame. It will simply be discarded.
*   `inplace`: If `True`, the DataFrame is modified in place; otherwise, a new DataFrame is returned with the reset index.
*   `col_level`: (Optional) If the columns have multiple levels (MultiIndex for columns), this determines which level the labels are inserted into. By default, it's the first level.
*   `col_fill`: (Optional) If `col_level` is not 0, defines how the other levels of the column MultiIndex are named. By default, it's an empty string.
*   `allow_duplicates`: (Optional) If `True`, allow new index values to be duplicates. The default is `False`.
*   `names`: (Optional) Used to name the new column(s) that result from the reset index. If `None`, the original index name(s) are used.

### `pandas.DataFrame.merge()`: Merge DataFrames

`DataFrame.merge()` is used to combine two DataFrames based on common columns or indices, similar to SQL joins. It's a fundamental operation for bringing together related data.

**Syntax:**

```python
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)
```

**Key Parameters:**

*   `left`: The DataFrame on the left side of the merge.
*   `right`: The DataFrame on the right side of the merge.
*   `how`: Specifies the type of merge to be performed. This is crucial as it determines which rows are kept:
    *   `'inner'` (default): Returns only the rows that have matching values in *both* DataFrames (intersection).
    *   `'left'`: Returns all rows from the left DataFrame, and matching rows from the right DataFrame. If there's no match, `NaN`s are filled for the right DataFrame's columns.
    *   `'right'`: Returns all rows from the right DataFrame, and matching rows from the left DataFrame. If there's no match, `NaN`s are filled for the left DataFrame's columns.
    *   `'outer'`: Returns all rows when there is a match in *either* the left or right DataFrame (union). Fills `NaN`s where there's no match.
    *   `'cross'`: Returns the Cartesian product of the two DataFrames, creating all possible pairs of rows.
*   `on`: (Optional) Column or list of column names to join on. These must be present in both DataFrames.
*   `left_on`: (Optional) Column or list of column names from the left DataFrame to use as keys.
*   `right_on`: (Optional) Column or list of column names from the right DataFrame to use as keys.

### Combining Data Sets with `pd.merge()`

**Standard Joins:**

Consider these two DataFrames:

**`adf`**
```
x1  x2
A   1
B   2
C   3
```

**`bdf`**
```
x1  x3
A   T
B   F
D   T
```

--- 

*   **Left Join (`how='left'`):**
    `pd.merge(adf, bdf, how='left', on='x1')`
    *Join matching rows from `bdf` to `adf`. Retains all rows from the left DataFrame (`adf`), and matching rows from `bdf`. If no match, `NaN`s are filled.*

    ```
    x1  x2   x3
    A   1.0  T
    B   2.0  F
    C   3.0  NaN
    ```

--- 

*   **Right Join (`how='right'`):**
    `pd.merge(adf, bdf, how='right', on='x1')`
    *Join matching rows from `adf` to `bdf`. Retains all rows from the right DataFrame (`bdf`), and matching rows from `adf`. If no match, `NaN`s are filled.*

    ```
    x1  x2   x3
    A   1.0  T
    B   2.0  F
    D   NaN  T
    ```

--- 

*   **Inner Join (`how='inner'`):**
    `pd.merge(adf, bdf, how='inner', on='x1')`
    *Join data. Retain only rows that have matching values in both sets.*

    ```
    x1  x2  x3
    A   1   T
    B   2   F
    ```

--- 

*   **Outer Join (`how='outer'`):**
    `pd.merge(adf, bdf, how='outer', on='x1')`
    *Join data. Retain all values and all rows from both sets. Fills `NaN`s where there's no match.*

    ```
    x1  x2   x3
    A   1.0  T
    B   2.0  F
    C   3.0  NaN
    D   NaN  T
    ```

### Filtering Joins

*   **Rows in `adf` that have a match in `bdf`:**
    `adf[adf.x1.isin(bdf.x1)]`

    ```
    x1  x2
    A   1
    B   2
    ```

--- 

*   **Rows in `adf` that do not have a match in `bdf`:**
    `adf[~adf.x1.isin(bdf.x1)]`

    ```
    x1  x2
    C   3
    ```

### `pandas.concat()`: Concatenate pandas objects

`pd.concat()` is a versatile function used to concatenate (join or stack) pandas objects (like Series or DataFrames) along a particular axis. It's often used to combine DataFrames either by adding rows (vertical stacking) or adding columns (horizontal joining).

**Syntax:**

```python
pd.concat(objs, axis=0, join='outer', ignore_index=False)
```

**Key Parameters:**

*   `objs`: A sequence (e.g., a list) of Series or DataFrame objects that you want to concatenate. ie a list of dataframe
*   `axis`: Determines along which axis the concatenation happens:
    *   `0` or `'index'` (default): Stacks objects vertically (adds rows).
    *   `1` or `'columns'`: Joins objects horizontally (adds columns).
*   `join`: How to handle indexes on the other axis when concatenating:
    *   `'outer'` (default): Uses the union of all indexes (keeps all index labels, filling `NaN`s where data is missing).
    *   `'inner'`: Uses the intersection of all indexes (keeps only shared index labels).
*   `ignore_index`: If `True`, a new default integer index (0, 1, 2, ...) is created for the resulting DataFrame. This is useful when the original indexes are not meaningful or would cause duplicates after concatenation.


**Key Use Cases:**

1.  **Append Rows (Stacking DataFrames Vertically):**
    *   **What it does:** This is the default behavior (`axis=0`). It stacks DataFrames one on top of the other, effectively adding more rows to your dataset.
    *   **Example:**
        ```python
        # df1 and df2 have the same columns
        combined_df = pd.concat([df1, df2])
        # Rows of df2 are appended to the end of df1
        ```

2.  **Append Columns (Joining DataFrames Horizontally):**
    *   **What it does:** By setting `axis=1`, `pd.concat()` places DataFrames side-by-side, adding more columns to your dataset. It aligns DataFrames based on their index.
    *   **Example:**
        ```python
        # df1 and df2 have the same index (or you want to align by index)
        combined_df = pd.concat([df1, df2], axis=1)
        # Columns of df2 are appended to the right of df1
        ```

### Pandas `apply()` Functions: Types and Syntax (Short)

Pandas offers powerful `apply()` methods to execute a function along an axis of a DataFrame or on elements of a Series. They are key for applying custom operations.

---

#### `DataFrame.apply()`: Apply a function along an axis

Applies a function to each column (`axis=0`) or each row (`axis=1`) of a DataFrame.

**Syntax:**

```python
df.apply(func, axis=0, args=(), **kwargs)
```

*   `func`: The function to apply.
*   `axis`: `0` or `'index'` for columns, `1` or `'columns'` for rows. (Default: `0`)
*   `args`, `kwargs`: Optional arguments to pass to `func`.

**Use Case:** Performing an operation that needs to look at an entire row or column at once (e.g., calculating a custom statistic for each column, or processing data based on multiple columns in a row).

---

#### `Series.apply()`: Apply a function to each element of a Series

Applies a function to each element in a Series.

**Syntax:**

```python
s.apply(func, args=(), **kwargs)
```

*   `func`: The function to apply.
*   `args`, `kwargs`: Optional arguments to pass to `func`.

**Use Case:** Element-wise operations (e.g., converting data types, cleaning strings, transforming values). Often faster than Python loops.

---

#### `GroupBy.apply()`: Apply a function to groups

After grouping data with `groupby()`, `apply()` performs a function independently on each group, then combines the results. This is highly flexible.

**Syntax:**

```python
df.groupby(by=...).apply(func, *args, **kwargs)
```

*   `func`: The function to apply to each sub-DataFrame (group).
*   `*args`, `**kwargs`: Optional arguments to pass to `func`.

**Use Case:** Performing complex group-wise transformations or aggregations that cannot be done with standard `agg()` methods (e.g., normalizing data within each group, fitting a model to each group).

### `ast.literal_eval()`: Safely Evaluate Python Literals

**What it does:**
`ast.literal_eval()` is a function in Python's `ast` module that can safely evaluate a string containing basic Python data structures (like lists, dictionaries, numbers, strings, booleans, `None`). It converts these strings into actual Python objects.

**Think of it like:** You have a string `"[1, 2, 'hello']"` and you want to turn it into a real Python list `[1, 2, 'hello']`. `ast.literal_eval()` does this.

**Why it's useful (and 'safe'):**
*   It's designed to be safer than Python's `eval()` function.
*   It can *only* process simple Python literals and structures. It **cannot** execute arbitrary Python code, import modules, or call functions.
*   This makes it suitable for parsing data from untrusted sources, where `eval()` would be a security risk.

**Limitations:**
*   It won't work if the string contains operators (like `+`, `-`), function calls, or any complex Python expressions.
*   While much safer than `eval()`, extremely large or maliciously crafted inputs can still cause resource exhaustion (like memory or CPU).

**In short:** Use it when you need to convert a string representation of basic Python data (like a list or dict) into the actual Python object, and you need to do it safely without running arbitrary code.

### `pandas.DataFrame.explode()`: Simple Explanation

**What it does:**
`df.explode()` takes a DataFrame column where each entry is a list (or similar collection), and transforms each item in those lists into its own separate row. All other column values in that row are duplicated.

**Think of it like:** If you have a row for a 'job' that has a list of 'skills' (e.g., `['Python', 'SQL']`), `explode()` will turn that one job row into two rows: one for 'Python' skill and one for 'SQL' skill, both linked back to the original job details.

ie , Transform each element of a list-like to a row, replicating index valuesx

**Key Parameters:**

*   `column`: The name of the column (or list of columns) that contains the list-like entries you want to 'explode'.
*   `ignore_index`: If `True`, the original index numbers will be discarded, and a new simple numerical index (0, 1, 2...) will be created for the resulting DataFrame. This is useful if the original index is no longer relevant after explosion.
