### `split()` Method Syntax

The `split()` method is used to split a string into a list of substrings. The general syntax is:

```python
string.split(separator, maxsplit)
```

*   `separator` (optional): The delimiter string to split by. If not specified, any whitespace is a separator, and empty strings are removed from the result.
*   `maxsplit` (optional): The maximum number of splits to do. If unspecified, it defaults to -1, meaning no limit on the number of splits.

This code first modifies `job_title` to "Doto Analyst" using `replace()`, and then uses `split(' ')` to split the string at each space, resulting in a list of two strings: `['Doto', 'Analyst']`.

````python
print('hello world'.split(' '))
````

````python
job_title = 'Data Analyst'
# Corrected to split by space
print(job_title.split(sep=' ', maxsplit=1))
````

The previous code `job_title.split(sep="", 1)` resulted in a `SyntaxError` because a positional argument (`1`) was placed after a keyword argument (`sep=""`).

To fix this, `maxsplit` should also be provided as a keyword argument, like `maxsplit=1`. The `job_title` variable was re-initialized to 'Data Analyst' for this example.

### Positional vs. Keyword Arguments

When you call a Python function, you pass arguments to it. These arguments can be passed in two main ways:

*   **Positional Arguments**: These are arguments whose value is determined by their position in the function call. The order in which they are passed matters.

    ```python
    # Example with positional arguments
    def greet(name, message):
        print(f"{message}, {name}!")

    greet("Alice", "Hello") # 'Alice' is assigned to name, 'Hello' to message
    ```

*   **Keyword Arguments**: These are arguments whose value is determined by explicitly naming the parameter they correspond to, using the `parameter=value` syntax. Their order generally does not matter (as long as all positional arguments come first).

    ```python
    # Example with keyword arguments
    def greet(name, message):
        print(f"{message}, {name}!")

    greet(name="Bob", message="Hi")
    greet(message="Greetings", name="Charlie") # Order doesn't matter here
    ```

In the `replace()` method, for instance, `job_title.replace('a', 'o', 2)` uses positional arguments for `oldvalue` ('a'), `newvalue` ('o'), and `count` (2). You can also use keyword arguments for clarity, like `job_title.replace(oldvalue='a', newvalue='o', count=2)`.

### Positional vs. Keyword Arguments with Examples

*   **`replace()` method:**
    *   `job_title.replace('a', 'o', 2)` uses **positional arguments**: `'a'` (oldvalue), `'o'` (newvalue), `2` (count).
    *   `job_title.replace(oldvalue='a', newvalue='o', count=2)` uses **keyword arguments**, improving readability.

*   **`split()` method:**
    *   `'Data Analyst'.split(' ', 1)` uses **positional arguments**: `' '` (separator), `1` (maxsplit).
    *   `'Data Analyst'.split(sep=' ', maxsplit=1)` uses **keyword arguments**.

This code demonstrates a classic way to swap two variable values: it uses a `temp_var` to temporarily hold one value while the other is assigned, then assigns the held value to the second variable.

````python
applicant_1 = "a"
applicant_2 = "b"

# Pythonic way to swap variables using tuple unpacking
applicant_1, applicant_2 = applicant_2, applicant_1

# Print the swapped values
print(applicant_1, applicant_2)
````

This code demonstrates the Pythonic way to swap two variable values directly using tuple unpacking. `applicant_1, applicant_2 = applicant_2, applicant_1` simultaneously assigns the original value of `applicant_2` to `applicant_1` and the original value of `applicant_1` to `applicant_2`, making the swap concise and efficient.

# **unpacking**

Assign each value in an iterable to a variable in a single statement

````python
job_skills = ['python', 'excel', 'sql','looker']
skill1, skill2, skill3,skill4 = job_skills
# this is unpacking

print(skill1)
print(skill2)
print(skill3)
print(skill4)
````

### Extended Unpacking with `*`

When you use `*variable_name` during unpacking, it allows you to assign a variable number of items from an iterable to a single variable, which becomes a list.

In the example `skill_concerned, *skill_dont_care = job_skills`:

*   `skill_concerned` takes the **first element** from `job_skills` ('python').
*   `*skill_dont_care` collects **all remaining elements** from `job_skills` into a new list (`['excel', 'sql', 'looker']`).

````python
job_skills = ['python', 'excel', 'sql', 'looker']

# Using extended unpacking
skill_concerned, *skill_dont_care = job_skills

print(skill_concerned)
print(skill_dont_care)
````

### Mutable vs Immutable in Python

In Python, objects are either **mutable** or **immutable** based on whether they can be altered after their creation.

*   **Mutable objects** can be changed after creation. Examples include:
    *   Lists
    *   Dictionaries
    *   Sets

*   **Immutable objects** cannot be changed after they are created. Examples include:
    *   Strings
    *   Integers
    *   Tuples

## Understanding the `for` loop

A `for` loop is used for iterating over a sequence (that is, an iterable) such as a list, tuple, dictionary, string, or range.

### Basic Syntax:

```python
for variable in iterable:
    # code to execute for each item
```

-   **`for`**: This keyword initiates the loop.
-   **`variable`**: This is a temporary variable that takes on the value of each item in the `iterable` during each iteration of the loop. You can name this variable anything you want (e.g., `item`, `element`, `num`, `char`).
-   **`in`**: This keyword specifies the `iterable` you want to loop through.
-   **`iterable`**: This is a collection of items that can be iterated over, one item at a time (e.g., a list `[1, 2, 3]`, a string `'hello'`, a `range(5)`).
-   **`:`**: A colon marks the end of the `for` statement header.
-   **Indented Block**: The code that follows, indented below the `for` statement, is executed once for each item in the `iterable`.

### What is an Iterable?

An **iterable** is any Python object capable of returning its members one by one. Essentially, it's something you can "iterate over" using a `for` loop.

Common examples of iterables include:

-   **Lists**: `[1, 2, 3]`
-   **Tuples**: `(1, 2, 3)`
-   **Strings**: `'hello'`
-   **Dictionaries**: `{'a': 1, 'b': 2}` (iterates over keys by default)
-   **Sets**: `{1, 2, 3}`
-   **`range()` objects**: `range(5)` (generates numbers from 0 to 4)
-   **Files**: When you open a file, you can iterate over its lines.

### Printing `number` inside the loop

When you print `number` (or whatever you named your loop variable) inside the loop, you are printing the **current item** that the loop is processing during that specific iteration.

In the output above, each line shows a single element from the `numbers` list, as `number` takes on the value of each element sequentially.

````python
numbers = [0, 1, 2, 3, 4, 5]

# syntax = for variable in iterable
# for number its a new variable we created to use in this statement
for number in numbers:
# numbers its the iterable we will be iterating through it
  print(number)


# In Python, an iterable is anything that you can loop over or 'iterate' through, such as a list, tuple, string, or range or anything
# The for loop takes items from the numbers list one by one until all items have been processed.
````

````python
numbers = [0, 1, 2, 3, 4, 5]

for number in numbers:
  print(number)
````

### Printing `numbers` inside the loop

If you print `numbers` (the iterable itself) inside the loop, you are printing the **entire list** (or iterable) during each iteration, not the individual item.

````python
numbers = [0, 1, 2, 3, 4, 5]

for number in numbers:
  print(numbers)
````

As you can see, the entire list `[0, 1, 2, 3, 4, 5]` is printed six times, once for each iteration of the loop. The `numbers` list itself doesn't change during the iteration, only the `number` variable does.

## Explanation of the Provided Code

```python
years = 0
index = 0
while years < 5:
 key, years = years_list[index]
 index += 1
print(key,years)
```

This code iterates through a list of (name, years_experience) pairs (`years_list`) until it finds an entry where the `years` of experience is 5 or greater. Let's go through it step by step:

1.  **`years = 0`**: Initializes a variable `years` to `0`. This variable will hold the years of experience from each person in the list and is also used as the loop's condition.
2.  **`index = 0`**: Initializes an `index` variable to `0`. This is used to access elements in `years_list` by their position.
3.  **`while years < 5:`**: This is the core of the loop. The code inside this `while` loop will continue to execute **as long as** the value of `years` is less than `5`. Once `years` becomes `5` or more, the loop will stop.
4.  **`key, years = years_list[index]`**: Inside the loop, this line unpacks a tuple from `years_list` at the current `index`. For example, if `years_list[0]` is `('Luke', 3)`, then `key` will become `'Luke'` and `years` will become `3`.
5.  **`index += 1`**: After processing an item, this line increments the `index` by `1`. This ensures that in the next iteration, the loop moves to the next item in `years_list`.
6.  **`print(key, years)`**: After the `while` loop finishes (i.e., when `years` is no longer less than `5`), this line prints the `key` and `years` from the *last* item processed by the loop that caused the condition `years < 5` to become `False`.

### How it works with your data:

Let's trace it with `years_list = [('Luke', 3), ('Kelly', 6), ('Ken', 4), ('Alex', 3), ('Mary', 9)]`:

-   **Initial**: `years = 0`, `index = 0`
-   **Iteration 1**:
    -   `years < 5` (0 < 5) is True.
    -   `key, years = years_list[0]` which is `('Luke', 3)`. So, `key` becomes `'Luke'`, `years` becomes `3`.
    -   `index` becomes `1`.
-   **Iteration 2**:
    -   `years < 5` (3 < 5) is True.
    -   `key, years = years_list[1]` which is `('Kelly', 6)`. So, `key` becomes `'Kelly'`, `years` becomes `6`.
    -   `index` becomes `2`.
-   **Loop Check**: Now `years` is `6`. The condition `years < 5` (6 < 5) is False. The loop terminates.
-   **After Loop**: `print(key, years)` will output: `Kelly 6` (these are the values from the last iteration before the loop terminated).

### List Comprehension Syntax

List comprehension provides a concise way to create lists in Python. It consists of a single line that combines a `for` loop and a `lambda` expression to process an iterable and construct a new list.

```python
[expression for item in iterable if condition]
```

*   **`expression`**: The item to be added to the new list (e.g., `item`, `item * 2`, `item.upper()`).
*   **`item`**: The variable representing each element from the `iterable`.
*   **`iterable`**: The sequence or collection you are looping over (e.g., a list, tuple, string, or `range()`).
*   **`if condition`** (optional): A filter that determines whether `item` is processed and included in the new list.

````python
# Example: Create a list of squares for numbers from 0 to 9
squares = [x**2 for x in range(10)]
print(f"List of squares: {squares}")

# Example: Create a list of even numbers from 0 to 9
even_numbers = [x for x in range(10) if x % 2 == 0]
print(f"List of even numbers: {even_numbers}")
````

Let's break down the code that filters `job_roles` based on `my_skills` using traditional `for` loops:

```python
qualified_roles = []

for job in job_roles:
    qualified = True

    for skill in my_skills:
        if skill not in job['skills']:
            qualified = False
            break

    if qualified:
        qualified_roles.append(job['role'])

print(qualified_roles)
```

Here's what each part does:

1.  **`qualified_roles = []`**:
    *   **Purpose**: This line initializes an empty list. This list will be used to store the names of the job roles that an applicant is qualified for (i.e., those that contain all `my_skills`).

2.  **`for job in job_roles:`**:
    *   **Purpose**: This is the **outer loop**. It iterates through each individual job dictionary present in the `job_roles` list. In each iteration, the variable `job` will hold one of these job dictionaries (e.g., `{'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']}`).

3.  **`qualified = True`**:
    *   **Purpose**: Inside the outer loop, *for each new job*, a boolean variable `qualified` is set to `True`. This acts as a flag. We optimistically assume the applicant is qualified for the current job until we find a reason (a missing skill) to prove otherwise.

4.  **`for skill in my_skills:`**:
    *   **Purpose**: This is the **inner loop**. For the current `job` being examined, this loop iterates through each skill listed in `my_skills` (which are the skills the applicant possesses, e.g., 'Python', 'SQL', 'Excel').

5.  **`if skill not in job['skills']:`**:
    *   **Purpose**: Inside the inner loop, this is a conditional check. For each `skill` from `my_skills`, it checks if that `skill` is *not* present in the `skills` list of the `current job`.
    *   **Example**: If `my_skills` has 'Python' and the `job['skills']` also has 'Python', this condition is `False`. If `my_skills` has 'Java' and `job['skills']` does *not* have 'Java', this condition is `True`.

6.  **`qualified = False`**:
    *   **Purpose**: If the condition `skill not in job['skills']` is `True` (meaning a required skill is missing), the `qualified` flag for the current job is immediately set to `False`.

7.  **`break`**:
    *   **Purpose**: Right after `qualified` is set to `False`, the `break` statement is executed. This immediately terminates the **inner loop** (`for skill in my_skills`). There's no need to check any further skills for this `job` because we already know the applicant is not qualified due to one missing skill.

8.  **`if qualified:`**:
    *   **Purpose**: After the inner loop finishes (either by checking all skills and finding them all present, or by `break`ing early because a skill was missing), this condition checks the final state of the `qualified` flag for the current `job`.

9.  **`qualified_roles.append(job['role'])`**:
    *   **Purpose**: If `qualified` is still `True` at this point (meaning all skills from `my_skills` were found in the `job['skills']`), then the `role` (title) of that `job` is added to the `qualified_roles` list.

10. **`print(qualified_roles)`**:
    *   **Purpose**: Finally, after the outer loop has processed all jobs in `job_roles`, this line prints the `qualified_roles` list, showing all the job titles for which the applicant is fully qualified.

### Simple Explanation of Loops and Logic

Let's break down the code's structure in simple terms:

1.  **Big Loop (`for job in job_roles:`)**:
    *   Imagine you have a stack of job applications (that's your `job_roles` list).
    *   This big loop picks up **one job application at a time**.
    *   For each job, it starts by saying, "Okay, for *this job*, I'm going to assume you *are* qualified (`qualified = True`)."

2.  **Small Loop (inside the Big Loop) (`for skill in my_skills:`)**:
    *   Now, still looking at that *one job application*, this small loop checks **each of *your* skills** (`my_skills`) one by one.
    *   **The `if` check inside the small loop (`if skill not in job['skills']:`):**
        *   It asks: "Is *this skill* of yours missing from what *this job* requires?"
        *   If the answer is **YES** (a skill is missing), it immediately changes its mind: "Oh, wait, you're **NOT** qualified for *this job* (`qualified = False`)."
        *   Then, `break` means: "Stop checking *your other skills* for *this specific job* right now; we already know you're not fully qualified for it. Let's move on."

3.  **After the Small Loop (but still for the same Big Loop job) (`if qualified:`)**:
    *   Once the small loop is done (either all your skills were checked, or it `break` because one was missing), it looks at the `qualified` flag for *this specific job*.
    *   It asks: "Did we ever find a reason to say you were `False` for *this job*?"
    *   If `qualified` is still `True` (meaning all your skills matched), then: "Great! Add *this job's title* to our list of `qualified_roles`."

4.  **The `print(qualified_roles)` Statement**:
    *   This `print` statement is **outside both loops**. It's only run **once**, right at the very end.
    *   It waits until the big loop has looked at *every single job application* and decided on qualification for all of them. Only then does it show you the final list of all the `qualified_roles`.

### How many `if` statements can you use inside a loop?

You can use **as many `if` statements as you need** inside any loop. Python doesn't have a limit on that. You can have `if`, `elif` (else if), and `else` blocks, and you can even nest `if` statements inside other `if` statements, as long as your indentation is correct and the logic makes sense for what you're trying to achieve.

### Explanation of the `qualified` variable

In the provided code, `qualified` is a boolean variable that acts as a **flag** to determine if a job seeker possesses all the required skills for a particular job role.

1.  **Initialization**: At the beginning of processing each new `job` in the `job_roles` list, `qualified` is set to `True`.
    ```python
    qualified = True
    ```
    This is an optimistic assumption: we initially assume the job seeker *is* qualified for the current job.

2.  **Conditional Update**: The code then iterates through `my_skills` (the skills the job seeker has). For each of these skills, it checks if that skill is *not* present in the `skills` list of the current `job`.
    ```python
    for skill in my_skills:
        if skill not in job['skills']:
            qualified = False
            break  # Stop checking skills for this job, we already know they're not fully qualified
    ```
    If even one skill from `my_skills` is found to be missing from the `job['skills']`, the `qualified` flag is immediately set to `False`, and the `break` statement exits the inner loop. There's no need to check further skills for that job once one is found missing.

3.  **Final Check**: After the inner loop completes (either by checking all skills or by breaking early), the value of `qualified` reflects whether the job seeker has *all* the necessary skills for that specific job.
    ```python
    if qualified:
        qualified_roles.append(job['role'])
    ```
    If `qualified` is still `True`, it means no missing skills were found, and thus the job role is added to the `qualified_roles` list.

### `ast` - Abstract Syntax Trees (Simplified)

The `ast` module in Python helps you work with the *structure* of your code itself. Think of it like this:

*   When you write Python code, it's just text.
*   Before Python can run that text, it first translates it into a detailed **tree-like structure** that represents all the commands, variables, and logic. This is called an **Abstract Syntax Tree (AST)**.

**What `ast` does:**

*   The `ast` module lets you **look at** and **manipulate** these internal ASTs.
*   You can use it to understand how Python interprets your code, find specific patterns in code, or even change the code's structure programmatically.

**In short**: It's a tool for advanced Python users who want to analyze, modify, or generate Python code by understanding its underlying structure rather than just treating it as text.

### `ast.literal_eval()` - Safely Evaluating Python Literals

`ast.literal_eval()` is a powerful function within the `ast` module. Its purpose is to **safely evaluate a string containing a Python literal structure** (like a string, number, list, dictionary, boolean, `None`, etc.) and convert it into the corresponding Python object.

*   **Safe Evaluation**: Unlike `eval()`, `ast.literal_eval()` is designed *not* to execute arbitrary Python code. It only processes Python's basic data structures. This makes it much safer when you need to convert a string (that looks like a Python list or dictionary) into an actual list or dictionary, especially if the string comes from an untrusted source.
*   **Use Case**: It's perfect for when you have a string like `"['Python', 'SQL']"` and you want to turn it into an actual Python list `['Python', 'SQL']`.

### Packages and Libraries in Python

In Python, the terms **library** and **package** are often used somewhat interchangeably, but there's a subtle distinction:

*   **Library**: A broader term referring to a collection of code (modules) that provides reusable functionality to solve specific problems. Think of it as a comprehensive toolkit. Python's Standard Library (e.g., `math`, `datetime`, `os`) is a prime example.

**In essence:** A package is a *type* of library, focusing on how the code is structured in your file system. All packages are libraries, but not all libraries are necessarily structured as multi-module packages (some might be just a single module, or a collection of scripts that aren't formally packaged).

## Python Terminology: Module, Package, and Library

*   **Module**: A single Python file (`.py`) containing code. Used for organizing code into smaller, reusable units.
*   **Package**: A directory of related modules, indicated by an `__init__.py` file. Helps structure larger projects and avoids naming conflicts.
*   **Library**: A broader term referring to a collection of packages or modules that provide reusable functionality. Offers pre-written code to simplify development.

These structures are essential for code reuse, organization, and maintaining large codebases efficiently.

 **Classes in Simple Terms**

A **class** is like a blueprint or a template for creating objects. It defines a set of characteristics (called **attributes**) and behaviors (called **methods**) that all objects created from that class will have.

A key aspect of a class is that it brings together both **data** (which we call 'attributes') and the **functions** (methods) that operate on that data into a single, cohesive unit. This bundling helps in creating more organized and reusable code.

Think of it this way:

*   **Blueprint (Class)**: "This is how you build a car. It needs wheels, an engine, and it can drive and stop."
*   **Actual Car (Object/Instance)**: "Here is *my* car. It's red, has a 2.0L engine, and I can drive it to work."

Classes allow you to group related data and functions together, making your code more organized, reusable, and easier to manage.

````python
class op_list:
    def __init__(self):
        """Initialize an empty list."""
        self._items = []

    def add(self, item):
        """Add an item to the end of the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the length of the list."""
        return len(self._items)
````

````python
class BaseSalary:
    def __init__(self, base_salary, bonus_rate=0.1, symbol="$"):
        """Initialize the Salary instance for managing salary computations.

        Args:
            base_salary (float): The base salary.
            bonus_rate (float): The bonus rate. Default is 0.1.
            symbol (str): The currency symbol. Default is "$".
        """
        self.base_salary = base_salary
        self.bonus_rate = bonus_rate
        self.symbol = symbol

    def __repr__(self):
        return f"{self.symbol}{self.base_salary:,.0f}"

    def calculate_total_salary(self):
        """Calculate the total salary based on the base salary and the bonus rate.

        Returns:
            float: The total salary.
        """
        total_salary = self.base_salary * (1 + self.bonus_rate)
        return total_salary

    def calculate_bonus(self):
        """Calculate the bonus based on the base salary and the bonus rate.

        Returns:
            float: The bonus amount.
        """
        bonus_amount = self.base_salary * self.bonus_rate
        return bonus_amount
````
