## `while` Loops - A Simple Explanation

A `while` loop keeps running a block of code again and again, **as long as a certain condition is true**.

Think of it like this:

"While this specific thing is true, keep doing these steps."

### How it works:

1.  **Check the Condition**: The loop first checks if its condition is `True` or `False`.
2.  **Do the Steps (if True)**: If the condition is `True`, the code inside the loop runs.
3.  **Check Again**: After the code runs, the loop goes back to step 1 and checks the condition again.
4.  **Stop (if False)**: If the condition becomes `False`, the loop stops, and the program moves on to whatever comes after the loop.

### Important:

You need to make sure that something inside your loop will eventually make the condition `False`. Otherwise, the loop will run forever (this is called an **infinite loop**).

### Simple Example:

```python
counter = 0
while counter < 3: # Condition: Is counter less than 3?
    print("Hello!")
    counter = counter + 1 # Change counter so condition eventually becomes False
print("Loop finished.")
```

**What happens:**
-   `counter` is 0. Is 0 < 3? Yes. Prints "Hello!". `counter` becomes 1.
-   `counter` is 1. Is 1 < 3? Yes. Prints "Hello!". `counter` becomes 2.
-   `counter` is 2. Is 2 < 3? Yes. Prints "Hello!". `counter` becomes 3.
-   `counter` is 3. Is 3 < 3? No. Loop stops.
-   Prints "Loop finished."

## Explanation of the Provided Code

```python
years = 0
index = 0
while years < 5:
 key, years = years_list[index]
 index += 1
print(key,years)
```

This code iterates through a list of (name, years_experience) pairs (`years_list`) until it finds an entry where the `years` of experience is 5 or greater. Let's go through it step by step:

1.  **`years = 0`**: Initializes a variable `years` to `0`. This variable will hold the years of experience from each person in the list and is also used as the loop's condition.
2.  **`index = 0`**: Initializes an `index` variable to `0`. This is used to access elements in `years_list` by their position.
3.  **`while years < 5:`**: This is the core of the loop. The code inside this `while` loop will continue to execute **as long as** the value of `years` is less than `5`. Once `years` becomes `5` or more, the loop will stop.
4.  **`key, years = years_list[index]`**: Inside the loop, this line unpacks a tuple from `years_list` at the current `index`. For example, if `years_list[0]` is `('Luke', 3)`, then `key` will become `'Luke'` and `years` will become `3`.
5.  **`index += 1`**: After processing an item, this line increments the `index` by `1`. This ensures that in the next iteration, the loop moves to the next item in `years_list`.
6.  **`print(key, years)`**: After the `while` loop finishes (i.e., when `years` is no longer less than `5`), this line prints the `key` and `years` from the *last* item processed by the loop that caused the condition `years < 5` to become `False`.

### How it works with your data:

Let's trace it with `years_list = [('Luke', 3), ('Kelly', 6), ('Ken', 4), ('Alex', 3), ('Mary', 9)]`:

-   **Initial**: `years = 0`, `index = 0`
-   **Iteration 1**:
    -   `years < 5` (0 < 5) is True.
    -   `key, years = years_list[0]` which is `('Luke', 3)`. So, `key` becomes `'Luke'`, `years` becomes `3`.
    -   `index` becomes `1`.
-   **Iteration 2**:
    -   `years < 5` (3 < 5) is True.
    -   `key, years = years_list[1]` which is `('Kelly', 6)`. So, `key` becomes `'Kelly'`, `years` becomes `6`.
    -   `index` becomes `2`.
-   **Loop Check**: Now `years` is `6`. The condition `years < 5` (6 < 5) is False. The loop terminates.
-   **After Loop**: `print(key, years)` will output: `Kelly 6` (these are the values from the last iteration before the loop terminated).

Let's break down the code that filters `job_roles` based on `my_skills` using traditional `for` loops:

```python
qualified_roles = []

for job in job_roles:
    qualified = True

    for skill in my_skills:
        if skill not in job['skills']:
            qualified = False
            break

    if qualified:
        qualified_roles.append(job['role'])

print(qualified_roles)
```

Here's what each part does:

1.  **`qualified_roles = []`**:
    *   **Purpose**: This line initializes an empty list. This list will be used to store the names of the job roles that an applicant is qualified for (i.e., those that contain all `my_skills`).

2.  **`for job in job_roles:`**:
    *   **Purpose**: This is the **outer loop**. It iterates through each individual job dictionary present in the `job_roles` list. In each iteration, the variable `job` will hold one of these job dictionaries (e.g., `{'role': 'Data Analyst', 'skills': ['Python', 'SQL', 'Excel']}`).

3.  **`qualified = True`**:
    *   **Purpose**: Inside the outer loop, *for each new job*, a boolean variable `qualified` is set to `True`. This acts as a flag. We optimistically assume the applicant is qualified for the current job until we find a reason (a missing skill) to prove otherwise.

4.  **`for skill in my_skills:`**:
    *   **Purpose**: This is the **inner loop**. For the current `job` being examined, this loop iterates through each skill listed in `my_skills` (which are the skills the applicant possesses, e.g., 'Python', 'SQL', 'Excel').

5.  **`if skill not in job['skills']:`**:
    *   **Purpose**: Inside the inner loop, this is a conditional check. For each `skill` from `my_skills`, it checks if that `skill` is *not* present in the `skills` list of the `current job`.
    *   **Example**: If `my_skills` has 'Python' and the `job['skills']` also has 'Python', this condition is `False`. If `my_skills` has 'Java' and `job['skills']` does *not* have 'Java', this condition is `True`.

6.  **`qualified = False`**:
    *   **Purpose**: If the condition `skill not in job['skills']` is `True` (meaning a required skill is missing), the `qualified` flag for the current job is immediately set to `False`.

7.  **`break`**:
    *   **Purpose**: Right after `qualified` is set to `False`, the `break` statement is executed. This immediately terminates the **inner loop** (`for skill in my_skills`). There's no need to check any further skills for this `job` because we already know the applicant is not qualified due to one missing skill.

8.  **`if qualified:`**:
    *   **Purpose**: After the inner loop finishes (either by checking all skills and finding them all present, or by `break`ing early because a skill was missing), this condition checks the final state of the `qualified` flag for the current `job`.

9.  **`qualified_roles.append(job['role'])`**:
    *   **Purpose**: If `qualified` is still `True` at this point (meaning all skills from `my_skills` were found in the `job['skills']`), then the `role` (title) of that `job` is added to the `qualified_roles` list.

10. **`print(qualified_roles)`**:
    *   **Purpose**: Finally, after the outer loop has processed all jobs in `job_roles`, this line prints the `qualified_roles` list, showing all the job titles for which the applicant is fully qualified.

### Simple Explanation of Loops and Logic

Let's break down the code's structure in simple terms:

1.  **Big Loop (`for job in job_roles:`)**:
    *   Imagine you have a stack of job applications (that's your `job_roles` list).
    *   This big loop picks up **one job application at a time**.
    *   For each job, it starts by saying, "Okay, for *this job*, I'm going to assume you *are* qualified (`qualified = True`)."

2.  **Small Loop (inside the Big Loop) (`for skill in my_skills:`)**:
    *   Now, still looking at that *one job application*, this small loop checks **each of *your* skills** (`my_skills`) one by one.
    *   **The `if` check inside the small loop (`if skill not in job['skills']:`):**
        *   It asks: "Is *this skill* of yours missing from what *this job* requires?"
        *   If the answer is **YES** (a skill is missing), it immediately changes its mind: "Oh, wait, you're **NOT** qualified for *this job* (`qualified = False`)."
        *   Then, `break` means: "Stop checking *your other skills* for *this specific job* right now; we already know you're not fully qualified for it. Let's move on."

3.  **After the Small Loop (but still for the same Big Loop job) (`if qualified:`)**:
    *   Once the small loop is done (either all your skills were checked, or it `break` because one was missing), it looks at the `qualified` flag for *this specific job*.
    *   It asks: "Did we ever find a reason to say you were `False` for *this job*?"
    *   If `qualified` is still `True` (meaning all your skills matched), then: "Great! Add *this job's title* to our list of `qualified_roles`."

4.  **The `print(qualified_roles)` Statement**:
    *   This `print` statement is **outside both loops**. It's only run **once**, right at the very end.
    *   It waits until the big loop has looked at *every single job application* and decided on qualification for all of them. Only then does it show you the final list of all the `qualified_roles`.

### How many `if` statements can you use inside a loop?

You can use **as many `if` statements as you need** inside any loop. Python doesn't have a limit on that. You can have `if`, `elif` (else if), and `else` blocks, and you can even nest `if` statements inside other `if` statements, as long as your indentation is correct and the logic makes sense for what you're trying to achieve.

## Module, Function, and Method - Simplified

Let's break down these terms with easy examples:

### 1. Function

*   **What it is**: A standalone block of code that does a specific job. You can call it anytime to perform that job.
*   **How to spot it**: It's defined using the `def` keyword, usually takes some inputs (arguments) in parentheses, and might `return` a result.

    **Simple Example:**

    ```python
    def greet(name):
        return f"Hello, {name}!"

    # Calling the function
    message = greet("Alice")
    print(message) # Output: Hello, Alice!
    ```

### 2. Method

*   **What it is**: A function that "belongs" to an object or a class. It performs an action specific to that object.
*   **How to spot it**: You call it using a dot (`.`) after an object's name. For example, `my_list.append()` or `my_dataframe.mean()`.

    **Simple Example:**

    ```python
    my_list = [1, 2, 3]

    # 'append' is a method of the list object
    my_list.append(4)
    print(my_list) # Output: [1, 2, 3, 4]

    my_string = "hello"
    # 'upper' is a method of the string object
    print(my_string.upper()) # Output: HELLO
    ```

### 3. Module

*   **What it is**: A file containing Python code. Think of it as a toolbox or a collection of related functions, methods, and other code.
*   **How to spot it**: It's a `.py` file. You bring its contents into your current program using the `import` keyword.

    **Simple Example:**

    Imagine you have a file named `calculator.py` with this content:

    ```python
    # calculator.py
    def add(a, b):
        return a + b

    def subtract(a, b):
        return a - b
    ```

    Now, in another Python script, you can use it like this:

    ```python
    import calculator # Importing the module

    # Using functions from the 'calculator' module
    result_sum = calculator.add(10, 5)
    print(result_sum) # Output: 15
    ```

## Clarifying Module, Method, and Function Relationships

Let's clear up how these concepts relate to each other:

*   **Function**: This is the most fundamental. A function is a standalone block of code designed to perform a specific task.
    *   *Example:* `def calculate_sum(a, b): return a + b`

*   **Method**: A method is essentially a function that *belongs to* an object or a class. It performs an action specific to that object's data.
    *   *Example:* `my_list.append(item)` - `append` is a method of the `list` object `my_list`.

*   **Module**: A module is a file ending in `.py` that contains Python code. This code can include functions, classes (which contain methods), variables, and other Python statements. It's a way to group related code together for organization and reuse.
    *   *Example:* A file named `my_utils.py` might contain several utility functions like `calculate_sum` and `format_string`.

*   **Package**: A package is a directory that contains multiple modules (and can also contain sub-packages). It's a way to organize related modules into a larger structure.
    *   *Example:* A `data_analysis` package might contain modules like `cleaning.py`, `visualization.py`, and `statistics.py`.

### In Summary:

*   A **function** is a piece of code that does something.
*   A **method** is a function that belongs to an object.
*   A **module** is a file that contains functions and/or classes (with methods).
*   A **package** is a folder that contains modules.

The . (dot) is used to access members of an object or a module.

When you use module.function_name, the . accesses a function inside that module.
When you use object.method_name, the . accesses a method associated with that object.
The () (parentheses) are used to call a function or a method.

When you see function_name(), you are telling Python to execute the code inside that function.
When you see object.method_name(), you are telling Python to execute the code inside that method, acting on that specific object.
